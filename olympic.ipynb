```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'http://olympics.cbc.ca/medals/'
html = requests.get(url).content
soup = BeautifulSoup(html,'lxml')
```

```python
tr = soup.find_all('tr',{'class':'or-ses-list or-js-sch-active'})
rank = []
country = []
gold = []
silver = []
bronze = []
total = []
for tr_i in tr:
    span = tr_i.find('span',{'class':'orp-rk'})
    rank.append(span.contents[0])
    
    span = tr_i.find('span',{'class':'or-team-n'})
    country.append(span['title'])
    
    td = tr_i.find('td',{'class':'or-md or-gold'})
    gold.append(td.contents[0])

    td = tr_i.find('td',{'class':'or-md or-silver'})
    silver.append(td.contents[0])
    
    td = tr_i.find('td',{'class':'or-md or-bronze'})
    bronze.append(td.contents[0])
    
    td = tr_i.find('td',{'class':'or-md or-total or-l'})
    total.append(td.contents[0])
```

```python
result = pd.DataFrame({'rank':rank,'country':country,'gold':gold,
                       'silver':silver,'bronze':bronze,'total':total},index=rank)
result = result[['country','gold','silver','bronze','total']]
```
